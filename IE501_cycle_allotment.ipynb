```python
import numpy as np
n = 4
m = 4
model = ConcreteModel()
model.X = Var(np.arange(1,n+1),np.arange(1,m+1),domain=Binary)
model.y1 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)
model.y2 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)
model.y3 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)
model.y4 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)
model.y5 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)
model.y6 = Var(np.arange(1,n+1), np.arange(1,m+1), domain=Binary)

model.constraints = ConstraintList()
```

```python
M=10000
for i in range(2,n):
    for j in range(2,m):
        model.constraints.add(
            model.y1[i,j]+model.y2[i,j]+model.y3[i,j]+model.y4[i,j]+ model.y5[i,j]+model.y6[i,j]==1
        )
        model.constraints.add(
            sum(model.X[i, k] for k in range(1, j )) + (j-1) * model.X[i, j] <= j - 1 + M * (1 - model.y1[i, j])
        )
        model.constraints.add(
            sum(model.X[i, k] for k in range(j + 1, m+1)) + (m - j) * model.X[i, j] <= (m - j) + M * (1 - model.y2[i, j])
        )
        model.constraints.add(
            sum(model.X[i + 1, k] for k in range(1, j + 1)) + j * model.X[i, j] <= j + M * (1 - model.y3[i, j])
        )
        model.constraints.add(
            sum(model.X[i + 1, k] for k in range(j, m+1)) + (m - j + 1) * model.X[i, j] <= (m - j + 1) + M * (1 - model.y4[i, j])
        )
        model.constraints.add(
            sum(model.X[i-1,k] for k in range(1,j+1)) + j*model.X[i,j] <= j +M*(1-model.y5[i,j])
        )

        model.constraints.add(
            sum(model.X[i-1,k] for k in range(j,m+1))+ (m-j+1)*model.X[i,j]<=m-j+1 + M*(1-model.y6[i,j])
        )
```

```python
for j in range(1,m+1):
  model.constraints.add(
            model.y1[1,j]+model.y2[1,j]+model.y3[1,j]+model.y4[1,j]==1
        )
```

```python
for j in range(1,m+1):
  model.constraints.add(
            sum(model.X[1, k] for k in range(1, j )) + (j - 1) * model.X[1, j] <= j - 1 + M * (1 - model.y1[1, j])
        )
  model.constraints.add(
      sum(model.X[1, k] for k in range(j + 1, m+1)) + (m - j) * model.X[1, j] <= (m - j) + M * (1 - model.y2[1, j])
  )
  model.constraints.add(
      sum(model.X[1 + 1, k] for k in range(1, j + 1)) + j * model.X[1, j] <= j + M * (1 - model.y3[1, j])
  )
  model.constraints.add(
      sum(model.X[1 + 1, k] for k in range(j, m+1)) + (m - j + 1) * model.X[1, j] <= (m - j + 1) + M * (1 - model.y4[1, j])
  )
```

```python
for j in range(1,m+1):
  model.constraints.add(
            model.y1[n,j]+model.y2[n,j]+model.y3[n,j]+model.y4[n,j]==1
        )
```

```python
for j in range(1,m+1):
  model.constraints.add(
            sum(model.X[n, k] for k in range(1, j )) + (j - 1) * model.X[n, j] <= j - 1 + M * (1 - model.y1[n, j])
        )
  model.constraints.add(
      sum(model.X[n, k] for k in range(j + 1, m+1)) + (m - j) * model.X[n, j] <= (m - j) + M * (1 - model.y2[n, j])
  )
  model.constraints.add(
      sum(model.X[n-1, k] for k in range(1, j + 1)) + j * model.X[n, j] <= j + M * (1 - model.y3[n, j])
  )
  model.constraints.add(
      sum(model.X[n-1, k] for k in range(j, m+1)) + (m - j + 1) * model.X[n, j] <= (m - j + 1) + M * (1 - model.y4[n, j])
  )
```

```python
model.pprint()
```

```python
model.obj = Objective(expr=sum(sum(model.X[i,j] for j in range(1,m+1))for i in range(1,n+1)),sense=maximize)
```
